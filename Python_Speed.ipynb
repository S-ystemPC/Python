## Unpack Operators / *args and **kwargs

```python
x = [1,12,23,232,213,123]
print(*x) #unpack operator
```

```python
x = [1,12,23,232,213,123]
print(*x) #unpack operator
print(x)
```

```python
def fun(x,y):
    print(x,y)

pairs = [ (0,1),(3,4) ]

for i in pairs:
    fun(i[0],i[1])
```

```python
def fun(x,y):
    print(x,y)

pairs = [(0,1),(3,4)]

for i in pairs:
    fun(*i)
```

```python
def fun(x,y):
    print(x,y)

fun(**{"y":7,"x":90})
```

```python
def fun(*args,**kwargs):
    print(args,kwargs)

fun(1,2,3,4,5,6,7,x=1,y=45)
```

```python
def fun(*args,**kwargs):
    print(*args)

fun(1,2,3,4,5,6,7,x=1,y=45)
```

## Scope and Globals

```python
x = "tim"

def fun(name):
    x = name

print(x)
fun("changed")
print(x)
```

```python
x = "tim"

def fun(name):
    global x   #avoid using this always because bad programmar practice
    x = name

print(x)
fun("changed")
print(x)
```

## Map & Filter

```python
x =[1,2,3,4,5,6,7,8,88,9,854,67,54]

mp = map(lambda i:i+2,x)
print(list(mp))
```

```python
x =[1,2,3,4,5,6,7,8,88,9,854,67,54]

mp = map(lambda i:i*2,x)
print(list(mp))
```

```python
x =[1,2,3,4,5,6,7,8,88,9,854,67,54]

mp = filter(lambda i:i%2==0,x)
print(list(mp))
```
