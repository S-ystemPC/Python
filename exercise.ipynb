#  Convert Celsius to Fahrenheit

```python
C = int(input("Enter Magnitude of Celsius: "))

def ctof():
    F = (C*1.8) + 32
    return F

magnitude_F = ctof()
print("The converted Value in Fahrenheit is",magnitude_F)
```

# Check Prime Number

```python
number = int(input('Enter your number here: '))

def is_prime(number):
    if number < 2:
        return False

    for i in range(2,number):
        if number % i == 0:
            return False

    return True

val = is_prime(number)
print(val)
```

# Even or Odd

```python
n = int(input('Enter number: '))

def check_even():
    if n % 2 == 0:
        print('Even')

    else:
        print('Odd')

check_even()
```

# Word Counter - like Google Docs

```python
txt = input('Enter text here: ')

def word_count():
    for i in range(1,len(txt)+1):
        if i == len(txt):
            print(i)

word_count()
```

# Print pyramid using *

```python
rows = 10
for i in range(1,rows+1):

    # ---------*
    # --------***
    # -------*****
    # ------*******
    # -----*********
    # ... and so on ...
    
    print(" "* (rows-i), end = "")
    print("*"* (2*i - 1))
```
