# Write a python programme to display even numbers between 1 to 10. Also print how many even numbers are there.

```python
count = 0
for i in range(1,10):
    if i%2 == 0:
        print (i)
        count += 1

print("We have",count,"even numbers")
```

#  Convert Celsius to Fahrenheit

```python
C = int(input("Enter Magnitude of Celsius: "))

def ctof():
    F = (C*1.8) + 32
    return F

magnitude_F = ctof()
print("The converted Value in Fahrenheit is",magnitude_F)
```
