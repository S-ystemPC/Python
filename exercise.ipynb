# Check Prime Number

```python
number = int(input('Enter your number here: '))

def is_prime(number):
    if number < 2:
        return False

    for i in range(2,number):
        if number % i == 0:
            return False

    return True

val = is_prime(number)
print(val)
```

# Calculate Factorial

```python
n = int(input('Enter the Number: '))

def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact*i
    
    print(fact)

factorial(n)
```

# Word Counter - like Google Docs

```python
txt = input('Enter text here: ')

def word_count():
    for i in range(1,len(txt)+1):
        if i == len(txt):
            print(i)

word_count()
```

# Print pyramid using *

```python
rows = 10
for i in range(1,rows+1):

    # ---------*
    # --------***
    # -------*****
    # ------*******
    # -----*********
    # ... and so on ...
    
    print(" "* (rows-i), end = "")
    print("*"* (2*i - 1))
```

# Check Palindrome String

```python
# A Palindrome String is same as Original String if Reversed
# For example: 1.racecar
#              2.abcddcba etc...

def is_palindrome(astring):
    astring = astring.lower()
    reverse_str = astring[::-1]
    if astring == reverse_str:
        return True
    else:
        return False

str = input("Enter your input here: ")
is_palindrome(str)
```
