```python
wt = int(input("Weight in grams: "))
unit = input("Kg or Lbs: ")
if unit.upper() == "K":
    converted = wt / 0.45
    print("Weight in Lbs: " + str(converted))
else:
    converted = wt * 0.45
    print("Weight in Kg: " + str(converted))
```
