## If Statement

```python
temperature = 40
if temperature > 35 :
    print("It is a Hot Day")
    print("Drink Plenty of water")
```

```python
temperature = 40
if temperature > 35:
    print("It's a Hot Day")
    print("Drink Plenty of water")
print("Done")
```

```python
temp = 45
if temp > 46:
    print("Harmful to Humans")
    print("Don't go outside")
elif temp > 40:
    print("Very Hot")
```

```python
temp = 45
if temp > 46:
    print("Harmful to Humans")
    print("Don't go outside")
elif temp > 40:
    print("Very Hot")
elif temp > 30:
    print("Hot")
else:
    print("Cold")
```

```python
wt = int(input("Weight in grams: "))
unit = input("Kg or Lbs: ")
if unit.upper() == "K":
    converted = wt / 0.45
    print("Weight in Lbs: " + str(converted))
else:
    converted = wt * 0.45
    print("Weight in Kg: " + str(converted))
```
