# Write a function

```python
def is_leap(year):
    leap = False
    if year%4==0:
        if year%100==0:
            if year%400==0:
                leap=True
            else:
                leap=False
        else:
            leap=True
    
    return leap

year = int(input())
print(is_leap(year))
```

# List Comprehensions

```python
if __name__ == '__main__':
    x = int(input())
    y = int(input())
    z = int(input())
    n = int(input())
    
    output = [[i,j,k] for i in range(0,x+1) for j in range(0,y+1) for k in range(0,z+1) if i+j+k != n]
    print(output)
```

# Find the Runner-Up Score!

```python
if __name__ == '__main__':
    n = int(input())
    arr = [int(i) for i in input().split()]
    
    maxm = arr[0]
    for score in arr:
        if score > maxm:
            maxm = score
            
    secm = -101
    for score in arr:
        if score > secm and score < maxm:
            secm = score
            
    print(secm)
```
